```python


"""
 Filters the merged F1 dataset based on optional parameters:
 season, driver full name, and constructor name."""
def filter_f1_data(df, season=None, driver=None, constructor=None):

        """Create a copy of to avoid modifiying orignal data """
filtered_df = df.copy()

      #This check rows that matches season value
if season is not None:
              filtered_df = filtered_df[filtered_df['season'] == season]

          #Combine forename and surname to match full driver name
if driver is not None:
              filtered_df = filtered_df[
                  (filtered_df['forename'] + " " + filtered_df['surname']).str.lower() == driver.lower()
              ]

      # Filters rows based on constructor name
if constructor is not None:
              filtered_df = filtered_df[filtered_df['name'].str.lower() == constructor.lower()]

return filtered_df


```
